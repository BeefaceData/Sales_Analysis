```python
import pandas as pd
import os
```

### Merging different data into a single csv file

```python
# df = pd.read_csv('./Sales_Data/Sales_April_2019.csv')
# df.head()
files = [file for file in os.listdir('./Sales_Data')]
all_months_data = pd.DataFrame()
for file in files:
    df = pd.read_csv('./Sales_Data/'+file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv('all_data.csv', index=False)
```

```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

### CLEANING UP THE DATA

```python
nan_df = all_data[all_data.isna().any(axis=1)] # check all the data with NaN
nan_df.head()
```

```python
all_data = all_data.dropna(how= "all") # to remove all the Data with NaN
all_data.head()
```

#### Find Or and delete it

```python
# temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or']
# temp_df.head()
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or'] # to find Or in the data and delete them
```

#### Convert these Columns to int type using pd.to_numeric

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

### Adding a Month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32') #Convert Month str to int
# all_data['Month'] = pd.to_datetime(all_data['Month'], format = '%m').dt.month_name() #Convert Month Number to Month Name
all_data.head()
```

### Adding a Sales Column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Adding a City column

```python
# using .apply()
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

### Best sales month


```python
best_sales = all_data.groupby('Month').sum(numeric_only=True)
best_sales.head()
```

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months, best_sales['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Months in Numbers')
plt.show()
```

### City with the best sales

```python
best_city = all_data.groupby('City').sum(numeric_only=True)
#best_city 
best_city = best_city[['Price Each', 'Month', 'Sales']] # To print selected columns
best_city
```

```python
import matplotlib.pyplot as plt
plot_city = all_data.groupby('City')
cities = [city for city, df in plot_city]
plt.bar(cities, best_city['Sales'])
plt.xticks(cities, rotation = 'vertical', size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('cities')
plt.show()
```

### What is the best time for Advertisement in hours

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date']) # split the order date into a date, time sequence
```

```python
# Add an hour and minute column

all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute

all_data.head()
```

```python
time_data = all_data[['Order ID', 'Order Date', 'Hour', 'Minute', 'Sales', 'City']]
time_data.head()
```

```python
hour = all_data.groupby('Hour')
hours = [hour for hour, df in hour]
plt.xticks(hours)
plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xlabel('Hours')
plt.ylabel('Number of orders')
plt.grid()
plt.show()

# reccomended time is 11am and 7pm
```

### Products that are often sold together

```python
all_data.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped'] = all_data.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID' ,'Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
for key, value in count.most_common(10):
    print(key, value)
```
